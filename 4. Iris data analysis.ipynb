```python
#Calculating basic stats of sepal length in cm
print("Mean of the sepal length is: ",dataframe['SepalLengthCm'].mean())
print("Median of the sepal length is: ",dataframe['SepalLengthCm'].median())
print("The Min of the sepal length is: ",dataframe['SepalLengthCm'].min())
print("The Max of the sepal length is: ",dataframe['SepalLengthCm'].max())
```

```python
#Calculating basic stats of sepal width in cm
print("Mean of the sepal width is: ",dataframe['SepalWidthCm'].mean())
print("Median of the sepal width is: ",dataframe['SepalWidthCm'].median())
print("The Min of the sepal width is: ",dataframe['SepalWidthCm'].min())
print("The Max of the sepal width is: ",dataframe['SepalWidthCm'].max())
```

```python
#Calculating basic stats of petal length in cm
print("Mean of the petal length is: ",dataframe['PetalLengthCm'].mean())
print("Median of the petal length is: ",dataframe['PetalLengthCm'].median())
print("The Min of the petal length is: ",dataframe['PetalLengthCm'].min())
print("The Max of the petal length is: ",dataframe['PetalLengthCm'].max())
```

```python
#Calculating basic stats of Petal Width in cm
print("Mean of the Petal Width is: ",dataframe['PetalWidthCm'].mean())
print("Median of the Petal Width is: ",dataframe['PetalWidthCm'].median())
print("The Min of the Petal Width is: ",dataframe['PetalWidthCm'].min())
print("The Max of the Petal Width is: ",dataframe['PetalWidthCm'].max())
```

```python
#Sorting the data frame by the "sepal length(cm)" column in ascending order
sorted_df_asc = dataframe.sort_values( by = 'SepalLengthCm')
sorted_df_asc
```

```python
#Sorting the data frame by the "sepal length(cm)" column in descending order
sorted_df_desc = dataframe.sort_values( by = 'SepalLengthCm', ascending = False)
sorted_df_desc
```

```python
#Select rows from the data frame where the "petal width(cm)" column is greater than 1.5
high_petal_width = dataframe.loc[dataframe['PetalWidthCm']>1.5].sort_values(by = 'PetalWidthCm')
high_petal_width
```

```python
#Select rows from the data frame where the "species" column contains the substring "versicolor"
versicolor_species = dataframe[dataframe['Species'].str.contains('versicolor')]
versicolor_species
```

```python
#If you simply want to know if data frame "species" column contains the substring "versicolor"
versicolor_species_bool = dataframe['Species'].str.contains('versicolor')
versicolor_species_bool
```

```python
#Group the data frame by the "species" column and calculate the mean values for all numeric columns for each species.
df_sp = dataframe.groupby('Species')
df_sp.first()
```

```python
print("Mean of SepalLengthCm is: ", df_sp['SepalLengthCm'].mean())
print("Mean of SepalWidthCm is: ", df_sp['SepalWidthCm'].mean())
print("Mean of PetalLengthCm is: ", df_sp['PetalLengthCm'].mean())
print("Mean of PetalWidthCm is: ", df_sp['PetalWidthCm'].mean())
```

```python
# Replace all occurrences of "iris" in the "species" column with "flower".
dataframe['Species'] = dataframe['Species'].replace(to_replace='Iris',value='flower')
df_iris = dataframe.loc[dataframe['Species']=="Iris"]
df_iris
```
